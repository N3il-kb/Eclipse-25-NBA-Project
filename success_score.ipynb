```python
#Gathers all advanced team ratings (ORTG, DRTG, NRTG) for the last 10 seasons
from nba_api.stats.endpoints import LeagueDashTeamStats
import pandas as pd
import time

def get_advanced_team_ratings(season):
    """Fetch advanced team ratings (ORTG, DRTG, NRTG) for a given season."""
    time.sleep(1)
    stats = LeagueDashTeamStats(
        season=season,
        season_type_all_star='Regular Season',
        measure_type_detailed_defense='Advanced'
    )
    df = stats.get_data_frames()[0]
    return df[['TEAM_NAME', 'OFF_RATING', 'DEF_RATING', 'NET_RATING']]

# Last 10 seasons (from current 2024-25 back to 2015-16)
seasons = [f"{y}-{str(y+1)[-2:]}" for y in range(2015, 2025)]

for season in seasons[::-1]:  # Reverse so newest is on top
    print(f"\n===== {season} Advanced Team Ratings =====\n")
    try:
        df = get_advanced_team_ratings(season)
        df_sorted = df.sort_values(by='NET_RATING', ascending=False)
        print(df_sorted.to_string(index=False))
    except Exception as e:
        print(f"⚠️ Could not fetch data for {season}: {e}")
```
